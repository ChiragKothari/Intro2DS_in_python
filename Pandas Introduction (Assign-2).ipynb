# Coursera Assignment-2, Part-1

## Olympics dataset

```python
import pandas as pd

df = pd.read_csv("D:\course1_downloads\course1_downloads\olympics.csv", index_col = 0, skiprows = 1)
```

```python
df.head()
```

```python
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns = {col : 'Gold' + col[4:]}, inplace = True)
    if col[:2] == '02':
        df.rename(columns = {col : 'Silver' + col[4:]}, inplace = True)
    if col[:2] == '03':
        df.rename(columns = {col : 'Bronze' + col[4:]}, inplace = True)
    if col[:2] == '№ ':
        df.rename(columns = {col : '#' + col[1:]}, inplace = True)
        
names_ids = df.index.str.split('\s\(')
df.index = names_ids.str[0]
df['ID'] = names_ids.str[1].str[:3]
df = df.drop('Totals')
df.head()
```

```python
def answer_zero():
    return df.iloc[0]
```

```python
answer_zero()
```

```python
df[df['Gold'] == df['Gold'].max()].index[0]
```

```python
def answer_one():
    return df[df['Gold'] == df['Gold'].max()].index[0]
```

```python
answer_one()
```

```python
df['diff'] = df['Gold'] - df['Gold.1']
df['diff'].idxmax()
```

```python
(df['Gold'] - df['Gold.1']).idxmax()
```

```python
def answer_two():
    return (df['Gold'] - df['Gold.1']).idxmax()
```

```python
answer_two()
```

```python
grt_df = df[(df['Gold'] > 0) & (df['Gold.1'] > 0)].copy()
grt_df['rel'] = grt_df['diff'] / grt_df['Gold.2']
grt_df['rel'].idxmax()
```

```python
def answer_three():
    grt_df = df[(df['Gold'] > 0) & (df['Gold.1'] > 0)].copy()
    grt_df['rel'] = grt_df['diff'] / grt_df['Gold.2']
    return grt_df['rel'].idxmax()
```

```python
answer_three()
```

```python
def answer_four():
    Points = df['Gold.2']*3 + df['Silver.2']*2 + df['Bronze.2']
    return Points
```

```python
answer_four()
```
